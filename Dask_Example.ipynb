```python
attributes_for_npi_lite = attributes_for_npi[['NPI','Provider Organization Name (Legal Business Name)','Provider Last Name (Legal Name)','CBSA','provider_taxonomy_desc']]

attributes_for_npi_lite['provider_taxonomy_desc'] = attributes_for_npi_lite['provider_taxonomy_desc'].fillna('None');

attributes_for_npi_lite=attributes_for_npi_lite.sort_values(['CBSA', 'provider_taxonomy_desc'])

attributes_for_npi_lite['CBSA']=attributes_for_npi_lite['CBSA'].fillna(0)

attributes_for_npi_lite['CBSA']=attributes_for_npi_lite['CBSA'].round(0).astype(int)

attributes_for_npi_lite.dtypes
```

```python
attributes_for_npi_lite.head(2)
```

# Choose CBSA and Specialty

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
@interact_manual(cbsa=attributes_for_npi_lite['CBSA'].unique(),desc=attributes_for_npi_lite['provider_taxonomy_desc'].unique())
def h(cbsa=[],desc=[]):
    selection_data=attributes_for_npi_lite[(attributes_for_npi_lite['CBSA']==cbsa) & (attributes_for_npi_lite['provider_taxonomy_desc']==desc)]
    selection = selection_data.merge(hop_teaming,left_on= 'NPI',right_on ='from_npi',how = 'left')
    
    results = pd.DataFrame(selection.groupby(['to_npi'])['patient_count'].sum().sort_values(ascending=False)).reset_index()
    results['to_npi'] = results['to_npi'].round(0).astype(int).astype(str)
    results['patient_count'] = results['patient_count'].round(0).astype(int)
    results.to_csv('results.csv')
    return results
```
